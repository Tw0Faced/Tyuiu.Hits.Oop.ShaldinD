```

//Интерфейс для действий студента
public interface IStudentActions
{
    void Register();       // Регистрация
    void Graduate();       // Выпуск
    void DisplaySummary(); // Краткая информация
}

//Интерфейс для уведомлений (для внедрения зависимости)
public interface INotificationService
{
    void SendMessage(string message);
}

//Реализация интерфейса уведомлений через консоль
public class ConsoleNotificationService : INotificationService
{
    public void SendMessage(string message)
    {
        Console.WriteLine($"\n[Уведомление]: {message}");
    }
}

//БАЗОВЫЙ КЛАСС STUDENT
public class Student : IStudentActions
{
    // Основные свойства
    public string Name { get; set; }
    public int Age { get; set; }
    public int Course { get; set; }
    public string Gender { get; set; }
    public string University { get; set; }
    public double AverageScore { get; set; }

    // Новые свойства
    public string Email { get; set; }
    public bool IsScholarshipHolder { get; set; }
    public int CreditsCompleted { get; set; }

    // Зависимость (сервис уведомлений)
    private readonly INotificationService _notificationService;

    // Конструктор с внедрением зависимости
    public Student(string name, int age, int course, string gender, string university, INotificationService notificationService)
    {
        Name = name;
        Age = age;
        Course = course;
        Gender = gender;
        University = university;
        _notificationService = notificationService;
    }

    // Методы базового класса
    public virtual void GetInfo()
    {
        Console.WriteLine($"\nИнформация о студенте:");
        Console.WriteLine($"Имя: {Name}");
        Console.WriteLine($"Возраст: {Age}");
        Console.WriteLine($"Пол: {Gender}");
        Console.WriteLine($"Курс: {Course}");
        Console.WriteLine($"Университет: {University}");
        Console.WriteLine($"Средний балл: {AverageScore}");
        Console.WriteLine($"Email: {Email}");
    }

    public virtual void Study()
    {
        // Старый вывод
        Console.WriteLine($"{Name} учится на курсе {Course}.");
        // Новый улучшенный вывод
        Console.WriteLine($"{Name} активно посещает пары и изучает материал на {Course}-м курсе.");
    }

    public virtual void TakeExam(string subject)
    {
        // Старый вывод
        Console.WriteLine($"{Name} сдаёт экзамен по предмету: {subject}");
        // Новый вывод
        Console.WriteLine($"{Name} готовится к экзамену по предмету {subject} в университете {University}.");
    }

    // Явная реализация интерфейса IStudentActions
    void IStudentActions.Register()
    {
        _notificationService.SendMessage($"{Name} успешно зарегистрирован(а) в системе.");
    }

    void IStudentActions.Graduate()
    {
        _notificationService.SendMessage($"{Name} окончил(а) университет {University}!");
    }

    void IStudentActions.DisplaySummary()
    {
        Console.WriteLine($"\nКраткая информация: {Name}, курс {Course}, университет {University}");
    }
}

// КЛАСС БАКАЛАВРА
public class BachelorStudent : Student
{
    public string Specialty { get; set; }
    public int InternshipHours { get; set; }
    public string Curator { get; set; }
    public string ResearchProject { get; set; }

    public BachelorStudent(string name, int age, int course, string gender, string university, string specialty, INotificationService notificationService)
        : base(name, age, course, gender, university, notificationService)
    {
        Specialty = specialty;
    }

    // Переопределение Study
    public override void Study()
    {
        Console.WriteLine($"{Name} изучает дисциплины по специальности: {Specialty}");
        Console.WriteLine($"{Name} активно участвует в учебных проектах и семинарах по {Specialty}.");
    }

    // Перегрузка TakeExam
    public void TakeExam(string subject, string form)
    {
        Console.WriteLine($"{Name} сдаёт экзамен по предмету {subject} в форме: {form}");
        Console.WriteLine($"{Name} выбрал(а) форму экзамена: {form}. Подготовка идёт полным ходом!");
    }

    // Новый метод
    public void AttendInternship()
    {
        Console.WriteLine($"{Name} проходит практику ({InternshipHours} часов).");
        Console.WriteLine($"Практика проходит под руководством куратора {Curator}.");
    }
}

// КЛАСС МАГИСТРАНТА
public class MasterStudent : Student
{
    public string Advisor { get; set; }
    public string ThesisTopic { get; set; }
    public int ResearchHours { get; set; }
    public string LabAccess { get; set; }

    public MasterStudent(string name, int age, int course, string gender, string university, string advisor, INotificationService notificationService)
        : base(name, age, course, gender, university, notificationService)
    {
        Advisor = advisor;
    }

    // Переопределение TakeExam
    public override void TakeExam(string subject)
    {
        Console.WriteLine($"{Name} сдаёт экзамен по {subject} под руководством {Advisor}");
        Console.WriteLine($"{Advisor} помогает {Name} подготовиться к защите магистерской работы.");
    }

    public void DefendThesis()
    {
        Console.WriteLine($"{Name} защищает магистерскую диссертацию на тему: {ThesisTopic}");
        Console.WriteLine($"{Name} выступает перед комиссией с темой: \"{ThesisTopic}\"");
    }
}

// КЛАСС АСПИРАНТА
public class PhDStudent : Student, IResearcher
{
    public string DissertationTopic { get; set; }
    public int Publications { get; set; }
    public bool IsTeachingAssistant { get; set; }

    public PhDStudent(string name, int age, int course, string gender, string university, string dissertation, INotificationService notificationService)
        : base(name, age, course, gender, university, notificationService)
    {
        DissertationTopic = dissertation;
    }

    // Переопределение Study
    public override void Study()
    {
        Console.WriteLine($"{Name} работает над диссертацией: {DissertationTopic}");
        Console.WriteLine($"{Name} проводит эксперименты и анализирует результаты по теме: {DissertationTopic}");
    }

    // Реализация интерфейса исследователя
    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит научное исследование.");
        Console.WriteLine($"{Name} работает в лаборатории над новой гипотезой.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} опубликовал статью. Общее количество публикаций: {++Publications}");
        Console.WriteLine($"{Name} представил результаты в научном журнале. Всего публикаций: {Publications}");
    }
}

// Интерфейс исследователя (множественное наследование)
public interface IResearcher
{
    void ConductResearch();
    void PublishPaper();
}


        INotificationService notifier = new ConsoleNotificationService();

        Student s1 = new BachelorStudent("Иван", 19, 2, "Мужской", "МГУ", "Информатика", notifier);
        Student s2 = new MasterStudent("Мария", 24, 5, "Женский", "СПбГУ", "Проф. Сидоров", notifier);
        Student s3 = new PhDStudent("Алексей", 28, 6, "Мужской", "НИУ ВШЭ", "Искусственный интеллект", notifier);

        s1.GetInfo();
        s1.Study();

        s2.GetInfo();
        s2.TakeExam("Математика");

        s3.GetInfo();
        s3.Study();

        IStudentActions actions = s1;
        actions.Register();
        actions.DisplaySummary();
        actions.Graduate();
```

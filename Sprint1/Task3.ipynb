```
public class Student
{
    private string name;
    private int age;
    private int course;

    public string Name
    {
        get { return name; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                name = value;
            else
                Console.WriteLine("Имя не может быть пустым!");
        }
    }

    public int Age
    {
        get { return age; }
        set
        {
            if (value > 0 && value < 120)
                age = value;
            else
                Console.WriteLine("Некорректный возраст!");
        }
    }

    public int Course
    {
        get { return course; }
        set
        {
            if (value > 0 && value <= 6)
                course = value;
            else
                Console.WriteLine("Некорректный курс!");
        }
    }

    public Student(string name, int age, int course)
    {
        Name = name;
        Age = age;
        Course = course;
    }

     public Student()
    {
        Name = "Неизвестный";
        Age = 18;
        Course = 1;
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"\n  Информация о студенте:");
        Console.WriteLine($"   Имя:   {Name}");
        Console.WriteLine($"   Возраст: {Age}");
        Console.WriteLine($"   Курс:   {Course}");
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится на курсе {Course}");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен");
    }

       public void AskQuestion(Student other)
    {
        Console.WriteLine($"{Name} задаёт вопрос студенту {other.Name}");
    }
}

class BachelorStudent : Student
{
    public string Specialty { get; set; }

    public BachelorStudent(string name, int age, int course, string specialty)
        : base(name, age, course)
    {
        Specialty = specialty;
    }

    public BachelorStudent() : base()
    {
        Specialty = "Не выбрана";
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} изучает дисциплины по специальности: {Specialty}");
    }

        public void AskMaster(MasterStudent master)
    {
        Console.WriteLine($"{Name} просит совета у магистранта {master.Name}");
    }
}

class MasterStudent : Student
{
    public string ScientificAdvisor { get; set; }

    public MasterStudent(string name, int age, int course, string advisor)
        : base(name, age, course)
    {
        ScientificAdvisor = advisor;
    }

      public MasterStudent() : base()
    {
        ScientificAdvisor = "Не назначен";
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен под руководством научного руководителя: {ScientificAdvisor}");
    }

    public void HelpPhD(PhDStudent phd)
    {
        Console.WriteLine($"{Name} помогает аспиранту {phd.Name} с исследованием");
    }
}

class PhDStudent : Student
{
    public string DissertationTopic { get; set; }

    public PhDStudent(string name, int age, int course, string topic)
        : base(name, age, course)
    {
        DissertationTopic = topic;
    }

    public PhDStudent() : base()
    {
        DissertationTopic = "Не определена";
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} работает над диссертацией: {DissertationTopic}");
    }

    public void DiscussWith(BachelorStudent bachelor)
    {
        Console.WriteLine($"{Name} обсуждает тему диссертации со студентом бакалавриата {bachelor.Name}");
    }
}

        Console.WriteLine("=========================================");
        Console.WriteLine("   Демонстрация студентов и наследия");
        Console.WriteLine("=========================================\n");
        Student s1 = new BachelorStudent("Иван", 19, 2, "Информатика");
        Student s2 = new MasterStudent("Мария", 23, 5, "Проф. Сидоров");
        Student s3 = new PhDStudent("Алексей", 27, 6, "Искусственный интеллект");

        s1.GetInfo();
        s1.Study();

        s2.GetInfo();
        s2.TakeExam();

        s3.GetInfo();
        s3.Study();
        
        Console.WriteLine("\n=========================================");
        Console.WriteLine("     Взаимодействие объектов");
        Console.WriteLine("=========================================\n");
        ((BachelorStudent)s1).AskMaster((MasterStudent)s2);
        ((MasterStudent)s2).HelpPhD((PhDStudent)s3);
        ((PhDStudent)s3).DiscussWith((BachelorStudent)s1);

        Console.WriteLine("\n=========================================");
        Console.WriteLine("     Конец демонстрации");
        Console.WriteLine("=========================================");


```

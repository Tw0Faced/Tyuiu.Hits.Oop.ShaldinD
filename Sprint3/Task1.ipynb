````

public class Vehicle
{
    public string Model { get; set; }

    public Vehicle(string model)
    {
        Model = model;
    }

    public virtual void Drive(int speed)
    {
        Console.WriteLine($"{Model} –¥–≤–∏–∂–µ—Ç—Å—è —Å–æ —Å–∫–æ—Ä–æ—Å—Ç—å—é {speed} –∫–º/—á.");
    }

    public void Drive(string mode)
    {
        Console.WriteLine($"{Model} –¥–≤–∏–∂–µ—Ç—Å—è –≤ —Ä–µ–∂–∏–º–µ: {mode}.");
    }

    public void Drive(bool isManual)
    {
        if (isManual)
            Console.WriteLine($"{Model}: –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ä—É—á–Ω–∞—è –∫–æ—Ä–æ–±–∫–∞ –ø–µ—Ä–µ–¥–∞—á.");
        else
            Console.WriteLine($"{Model}: –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–æ–±–∫–∞ –ø–µ—Ä–µ–¥–∞—á.");
    }
}

public class Car : Vehicle
{
    public Car(string model) : base(model) { }

    public override void Drive(int speed)
    {
        base.Drive(speed);
        Console.WriteLine($"{Model}: –∞–∫—Ç–∏–≤–∏—Ä–æ–≤–∞–Ω–∞ —Å–∏—Å—Ç–µ–º–∞ —Å—Ç–∞–±–∏–ª–∏–∑–∞—Ü–∏–∏ –Ω–∞ —Å–∫–æ—Ä–æ—Å—Ç–∏ {speed} –∫–º/—á.");
    }
}

public class Motorcycle : Vehicle
{
    public Motorcycle(string model) : base(model) { }

    public override void Drive(int speed)
    {
        Console.WriteLine($"{Model} –ª–µ—Ç–∏—Ç –ø–æ –¥–æ—Ä–æ–≥–µ —Å–æ —Å–∫–æ—Ä–æ—Å—Ç—å—é {speed} –∫–º/—á! üèçÔ∏è");
    }
}

Vehicle vehicle = new Vehicle("Generic Vehicle");
Car car = new Car("BMW");
Motorcycle moto = new Motorcycle("Yamaha");

vehicle.Drive(70);
car.Drive(120);
moto.Drive(160);

Console.WriteLine();

car.Drive("Eco");
car.Drive(true);
````

````
public class Animal
{
    private string _name;
    private int _age;

    public string Name
    {
        get => _name;
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                _name = "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ";
        }
    }

    public int Age
    {
        get => _age;
        set
        {
            if (value >= 0 && value <= 100)
                _age = value;
            else
                _age = 0;
        }
    }

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} –∏–∑–¥–∞–µ—Ç –∑–≤—É–∫.");
    }

    public void MakeSound(int volume)
    {
        Console.WriteLine($"{Name} –∏–∑–¥–∞–µ—Ç –∑–≤—É–∫ —Å –≥—Ä–æ–º–∫–æ—Å—Ç—å—é {volume} –¥–ë.");
    }

    public void MakeSound(string sound)
    {
        Console.WriteLine($"{Name} –≥–æ–≤–æ—Ä–∏—Ç: {sound}");
    }
}

public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} –ª–∞–µ—Ç: –ì–∞–≤-–≥–∞–≤!");
    }
}

public class Cat : Animal
{
    public bool LikesMilk { get; set; }

    public Cat(string name, int age, bool likesMilk) : base(name, age)
    {
        LikesMilk = likesMilk;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} –º—è—É–∫–∞–µ—Ç: –ú—è—É!");
    }
}

public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} –ø–æ–µ—Ç: –ß–∏—Ä–∏–∫-—á–∏—Ä–∏–∫!");
    }
}

public class Fish : Animal
{
    public bool IsSaltwater { get; set; }

    public Fish(string name, int age, bool isSaltwater) : base(name, age)
    {
        IsSaltwater = isSaltwater;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} –Ω–µ –∏–∑–¥–∞–µ—Ç –∑–≤—É–∫–æ–≤ ‚Äî –æ–Ω–∞ –ø—Ä–æ—Å—Ç–æ –ø—É–∑—ã—Ä–∏—Ç—Å—è");
    }
}

var animals = new List<Animal>
{
    new Dog("–ë–æ–±–∏–∫", 5, "–û–≤—á–∞—Ä–∫–∞"),
    new Cat("–ú—É—Ä–∫–∞", 3, true),
    new Bird("–ö–µ—à–∞", 2, 0.4),
    new Fish("–ù–µ–º–æ", 1, true)
};

foreach (var animal in animals)
{
    animal.MakeSound();
}

Console.WriteLine("\n–ü–µ—Ä–µ–≥—Ä—É–∑–∫–∞ –º–µ—Ç–æ–¥–æ–≤:");
animals[0].MakeSound(80);
animals[1].MakeSound("–ú—Ä—Ä—Ä...");
````

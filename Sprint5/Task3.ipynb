```
using System;
using System.Collections.Generic;

public delegate void LogHandler(string message);

public delegate void ArticlePublished(string title);

public delegate void ExperimentAdded(string exp);

public delegate void PaperPublished();

public interface IWriter
{
    void PublishArticle(string title);
}

public interface IResearcher
{
    void AddExperiment(string exp);
    void PublishPaper();
}


public class NotificationService
{
    public LogHandler Logger;

    public NotificationService()
    {
        Logger = Console.WriteLine;
    }

    public void Notify(string message)
    {
        Logger?.Invoke($"[Уведомление] {message}");
    }
}

public class Student
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string University { get; set; }
    public double Rating { get; set; }

    public Student(string name, int age, string university, double rating)
    {
        Name = name;
        Age = age;
        University = university;
        Rating = rating;
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} изучает материалы.");
    }

    public virtual void AttendLecture()
    {
        Console.WriteLine($"{Name} посещает лекцию.");
    }

    public virtual void ShowInfo()
    {
        Console.WriteLine($"{Name}, {Age} лет, {University}. Рейтинг: {Rating}");
    }
}

public class Programmer : Student, IWriter
{
    private readonly NotificationService _notify;

    public List<string> Articles { get; } = new List<string>();

    public event ArticlePublished OnArticlePublished;

    public Programmer(string name, int age, string uni, double rating, NotificationService notify)
        : base(name, age, uni, rating)
    {
        _notify = notify;
    }

    void IWriter.PublishArticle(string title)
    {
        Articles.Add(title);
        OnArticlePublished?.Invoke(title);

        _notify.Notify($"Программист {Name} опубликовал статью: {title}");
    }

    public void Code()
    {
        Console.WriteLine($"{Name} пишет код...");
    }
}

public class Biologist : Student, IResearcher
{
    private readonly NotificationService _notify;

    public List<string> Experiments { get; } = new List<string>();

    public event ExperimentAdded OnExperimentAdded;
    public event PaperPublished OnPaperPublished;

    public Biologist(string name, int age, string uni, double rating, NotificationService notify)
        : base(name, age, uni, rating)
    {
        _notify = notify;
    }

    void IResearcher.AddExperiment(string exp)
    {
        Experiments.Add(exp);
        OnExperimentAdded?.Invoke(exp);
        _notify.Notify($"Биолог {Name} добавил эксперимент: {exp}");
    }

    void IResearcher.PublishPaper()
    {
        OnPaperPublished?.Invoke();
        _notify.Notify($"Биолог {Name} опубликовал научную работу.");
    }

    public void Research()
    {
        Console.WriteLine($"{Name} исследует образцы...");
    }
}

        var notify = new NotificationService();

        Programmer s2 = new Programmer("Максим", 21, "ИТМО", 4.7, notify);

        s2.OnArticlePublished += title =>
            Console.WriteLine($"[LOG] Статья добавлена: {title}");

        ((IWriter)s2).PublishArticle("Нейросети и оптимизация");

        Biologist s3 = new Biologist("Анна", 22, "МГУ", 4.9, notify);

        s3.OnExperimentAdded += exp =>
            Console.WriteLine($"[LOG] Эксперимент добавлен: {exp}");

        s3.OnPaperPublished += () =>
            Console.WriteLine("[LOG] Научная работа опубликована");

        ((IResearcher)s3).AddExperiment("Тестирование ML-модели");
        ((IResearcher)s3).PublishPaper();


        Console.WriteLine("\n==== ГРУППА СТУДЕНТОВ ====");
        List<Student> group = new List<Student>() { s2, s3 };

        foreach (var st in group)
        {
            st.ShowInfo();
            st.Study();
        }
```

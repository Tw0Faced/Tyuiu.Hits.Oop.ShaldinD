```
public class Student
{
    public string Name { get; set; }
    public int Age { get; set; }
    public int Course { get; set; }
    public string Gender { get; set; }
    public string University { get; set; }
    public double AverageScore { get; set; }

    public Student(string name, int age, int course, string gender, string university, double averagescore)
    {
        Name = name;
        Age = age;
        Course = course;
        Gender = gender;
        University = university;
        AverageScore = averagescore;
    }

    public virtual void GetInfo()
    {
        Console.WriteLine(new string('-', 40));
        Console.WriteLine($"Имя: {Name}");
        Console.WriteLine($"Возраст: {Age}");
        Console.WriteLine($"Пол: {Gender}");
        Console.WriteLine($"Курс: {Course}");
        Console.WriteLine($"Университет: {University}");
        Console.WriteLine($"СреднийБалл: {AverageScore:F2}");
        Console.WriteLine(new string('-', 40));
    }

    public virtual void Study() => Console.WriteLine($"{Name} учится.");
    public virtual void TakeExam() => Console.WriteLine($"{Name} сдаёт экзамен.");
    public void Introduce() => Console.WriteLine($"Привет, меня зовут {Name}, я учусь в {University}.");
    public void ImproveAvarageScore(double value) => AverageScore += value;
    public void ChangeUniversity(string newUniversity) => University = newUniversity;
}

public interface IResearcher
{
    void Research();
}

public interface ITeacher
{
    void Teach();
}


// Простое наследование
class BachelorStudent : Student
{
    public string Specialty { get; set; }
    public string InternshipPlace { get; set; }
    public string Club { get; set; }
    public string ProjectTitle { get; set; }

    public BachelorStudent(string name, int age, int course, string gender, string university, double averagescore, string specialty)
        : base(name, age, course, gender, university, averagescore)
    {
        Specialty = specialty;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} изучает дисциплины по специальности: {Specialty}");
    }

    public void DoInternship() => Console.WriteLine($"{Name} проходит практику в {InternshipPlace}");
    public void JoinClub(string clubName)
    {
        Club = clubName;
        Console.WriteLine($"{Name} вступил в клуб {Club}");
    }
    public void DefendProject() => Console.WriteLine($"{Name} защищает проект: {ProjectTitle}");
}

// Сложное наследование
class MasterStudent : Student, IResearcher
{
    public string ScientificAdvisor { get; set; }
    public string ThesisTitle { get; set; }
    public double Scholarship { get; set; }
    public string ResearchArea { get; set; }

    public MasterStudent(string name, int age, int course, string gender, string university, double averagescore, string advisor)
        : base(name, age, course, gender, university, averagescore)
    {
        ScientificAdvisor = advisor;
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен под руководством научного руководителя: {ScientificAdvisor}");
    }

    public void DefendThesis() => Console.WriteLine($"{Name} защищает магистерскую работу: {ThesisTitle}");
    public void ApplyForGrant() => Console.WriteLine($"{Name} подаёт заявку на грант.");
    public void PublishArticle(string journal) => Console.WriteLine($"{Name} публикует статью в журнале {journal}");
    public void Research() => Console.WriteLine($"{Name} проводит исследование в области {ResearchArea}");
}

// Множественное наследование (через интерфейсы)
class PhDStudent : Student, IResearcher, ITeacher
{
    public string DissertationTopic { get; set; }
    public int PublicationsCount { get; set; }
    public string TeachingAssistant { get; set; }
    public string Conference { get; set; }

    public PhDStudent(string name, int age, int course, string gender, string university, double averagescore, string topic)
        : base(name, age, course, gender, university, averagescore)
    {
        DissertationTopic = topic;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} работает над диссертацией: {DissertationTopic}");
    }

    public void TeachStudents() => Console.WriteLine($"{Name} ведёт занятия для студентов.");
    public void PresentAtConference() => Console.WriteLine($"{Name} выступает на конференции: {Conference}");
    public void WriteArticle(string title) => Console.WriteLine($"{Name} написал статью: {title}");

    public void Research() => Console.WriteLine($"{Name} исследует тему: {DissertationTopic}");
    public void Teach() => Console.WriteLine($"{Name} преподаёт как ассистент.");
}
 BachelorStudent b = new BachelorStudent("Иван", 19, 2, "М", "МГУ", 4.2, "Информатика");
        b.InternshipPlace = "Яндекс";
        b.ProjectTitle = "Система рекомендаций";
        b.GetInfo();
        b.Study();
        b.DoInternship();
        b.DefendProject();

        Console.WriteLine();

        MasterStudent m = new MasterStudent("Мария", 23, 5, "Ж", "СПбГУ", 4.6, "Проф. Сидоров");
        m.ResearchArea = "Машинное обучение";
        m.ThesisTitle = "Глубокие нейросети";
        m.GetInfo();
        m.TakeExam();
        m.Research();
        m.DefendThesis();

        Console.WriteLine();

        PhDStudent p = new PhDStudent("Алексей", 27, 6, "М", "МИФИ", 4.9, "Искусственный интеллект");
        p.Conference = "AI-2025";
        p.GetInfo();
        p.Study();
        p.Research();
        p.Teach();
        p.PresentAtConference();
```
